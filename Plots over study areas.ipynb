```python
import xarray as xr
import matplotlib.pyplot as plt
```

```python
hindcast_observation = xr.open_zarr('D:\Downloads\s2s-ai-challenge-template-hydroinformatics-group-master-data\data\hindcast-like-observations_2000-2019_biweekly_deterministic.zarr')
```

# Spain (Barcelona and Murcia)

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15,6))

Barcelona_tp = hindcast_observation.tp.isel(lead_time = 0).sel(longitude = 1.5, latitude = 41.4, method = 'nearest')
Barcelona_tp.plot(ax=axes[0])

Brcelona_tp_yearly_sum = Barcelona_tp.resample(forecast_time='Y').sum(dim = 'forecast_time')
Brcelona_tp_yearly_sum.plot(ax=axes[1])
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15,6))

Barcelona_t2m = hindcast_observation.t2m.isel(lead_time = 0).sel(longitude = 1.5, latitude = 41.4, method = 'nearest')
Barcelona_t2m.plot(ax = axes[0])

Barcelona_t2m_yearly_mean = Barcelona_t2m.resample(forecast_time='3M').mean(dim = 'forecast_time')
Barcelona_t2m_yearly_mean.plot(ax = axes[1])
```
